```python

from time import time
import pandas as pd
def etl_nyc_taxi(chunk_size, connection):
    total_rows = 1000000  # Number of rows to process
    
    # Define the starting row and chunk index
    start_row = 0
    chunk_index = 0
    
    df_data = pd.read_csv('yellow_tripdata_2016-02.csv', chunksize=chunk_size)
    num_iterations = total_rows // chunk_size  # Calculate the number of iterations
    count = 0
    overall_start_time = time()

    for chunk in df_data:
        t_start = time()
        chunk['tpep_pickup_datetime'] = pd.to_datetime(chunk['tpep_pickup_datetime'])
        chunk['tpep_dropoff_datetime'] = pd.to_datetime(chunk['tpep_dropoff_datetime'])
        chunk.to_sql(name='nyc_yellow_taxi', con=connection, if_exists='append')
        count += 1
        t_end = time()
        print(f'{count}) loaded next chunk {t_end - t_start:.3f} seconds')
        if count >= num_iterations:
            break  # Stop the loop once the desired number of iterations is reached

    overall_end_time = time() - overall_start_time
    print(f'finished loading a total of {chunk_size * count:,} records in {overall_end_time:.3f} seconds')
```

```python
etl_nyc_taxi(100000,engine)
```

```python
#The average trip amount given by passengers

avg_trip_amount_query = '''
    SELECT
        AVG(fare_amount / passenger_count) AS avg_trip_amount_per_passenger
    FROM
        nyc_yellow_taxi
    WHERE
        passenger_count > 0 AND fare_amount > 0; -- To avoid division by zero and exclude zero fares
'''

pd.read_sql(avg_trip_amount_query, engine)
```

```python
# The average trip distance by passengers

avg_trip_distance_query = '''
    SELECT
        AVG(trip_distance / passenger_count) AS avg_trip_distance_per_passenger
    FROM
        nyc_yellow_taxi
    WHERE
        passenger_count > 0; -- To avoid division by zero
'''

pd.read_sql(avg_trip_distance_query, engine)


```

```python
# Number of trips flagged as 'store and forward'

store_and_forward_query = '''
    SELECT
        SUM(CASE WHEN store_and_fwd_flag = 'Y' THEN 1 ELSE 0 END) AS store_and_forward_count
    FROM
        nyc_yellow_taxi;
'''
pd.read_sql(store_and_forward_query, engine)
```

```python
report = '''


SELECT 
    CURRENT_DATE AS ingestion_date,

    -- Total trips
    (SELECT COUNT(*) FROM nyc_yellow_taxi) AS total_trips,

    -- Average trip distance
    (SELECT AVG(trip_distance) FROM nyc_yellow_taxi) AS avg_trip_distance,
    
    -- trips flagged as store and forward
    (SELECT
        SUM(CASE WHEN store_and_fwd_flag = 'Y' THEN 1 ELSE 0 END) AS store_and_forward_count
     FROM
        nyc_yellow_taxi) AS store_and_forward,

    -- Vendor with the most trips
    (SELECT "VendorID"
     FROM (
         SELECT "VendorID", COUNT(*) AS trip_count
         FROM nyc_yellow_taxi
         GROUP BY "VendorID"
         ORDER BY trip_count DESC
         LIMIT 1
     ) AS most_trips) AS vendor_most_trips,

    -- Vendor with the least trips
    (SELECT "VendorID"
     FROM (
         SELECT "VendorID", COUNT(*) AS trip_count
         FROM nyc_yellow_taxi
         GROUP BY "VendorID"
         ORDER BY trip_count ASC
         LIMIT 1
     ) AS least_trips) AS vendor_least_trips,

    -- Average passenger count
    (SELECT AVG(passenger_count) FROM nyc_yellow_taxi) AS avg_passenger_count,

    -- Average trip amount per passenger
    (SELECT AVG(fare_amount / passenger_count)
     FROM nyc_yellow_taxi
     WHERE passenger_count > 0 AND fare_amount > 0) AS avg_trip_amount_per_passenger,

    -- Average trip distance per passenger
    (SELECT AVG(trip_distance / passenger_count)
     FROM nyc_yellow_taxi
     WHERE passenger_count > 0) AS avg_trip_distance_per_passenger,

    -- Shared ride count
    (SELECT COUNT(*)
     FROM nyc_yellow_taxi
     WHERE passenger_count > 1) AS shared_ride_count,

    -- Average fare amount
    (SELECT AVG(fare_amount) FROM nyc_yellow_taxi) AS average_fare_amount,

    -- Total revenue
    (SELECT SUM(tolls_amount + improvement_surcharge) FROM nyc_yellow_taxi) AS total_revenue,

    -- Average total amount
    (SELECT AVG(total_amount) FROM nyc_yellow_taxi) AS average_total_amount;
'''
# Execute the report query
result_df = pd.read_sql(report, engine)



# Display the result
print(result_df)
```

```python
def transform_data(query,connection):
    try:
        report_query = pd.read_sql(query,connection)
        print('successfully transformed the data')
        return report_query      
    except:
        print('Encountered error while transforming')
```

```python
def load_report_to_warehouse(dataframe,table,connection):
    try:
        dataframe.to_sql(table, con=connection,if_exists='append')
        print(f'successfully updated {table} table')
    except:
        print(f'Could not update {table}')
```
